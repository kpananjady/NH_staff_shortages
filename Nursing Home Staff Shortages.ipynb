# What this file does

It looks at CMS Nursing Home data to plot the % of Nursing Homes reporting that they have staff shortages each week. The file looks at Nursing Staff, Clinical Staff, Aides and Other Staff separately, and colors resulting % bars by month.

```python
import pandas as pd
```

```python
df = pd.read_csv('CMS_COVID-19_Nursing_Home_Dataset (4).csv')
df['Week Ending'] = pd.to_datetime(df['Week Ending'])
df['month'] = df['Week Ending'].dt.month
```
